```python
hw = [
    (56, 120),
    (14, 30),
]
H, W = 25, 25

preds, atts = cvt(imgs, rots, trans, intrins, extrins, post_rots, post_trans, return_att=True)

print(atts.shape)
rearranged_atts = []
for i, att in enumerate(atts):
    rearrangd_att = rearrange(att, '(b m) (H W) (n h w) -> b m H W n h w', n=6, m=4, H=H, W=W, h=hw[i][0],
                              w=hw[i][1])
    rearranged_atts.append(rearrangd_att)

# Visualization
att = rearranged_atts[-1]  # get the last attention layer
att = att[0].detach()  # batch size is 1, we remove the batch dim here.
mean_att = torch.mean(att, dim=0)
preds = preds.softmax(dim=1)
pred, labels = save_pred(preds, labels, './')
```

```python
bboxes = [
    (6, 19, 7, 20),
]  # (y1, x1, y2, x2)
# change this to an actual, meaningful bbox

cam_to_bev_atts = []
for i, att in enumerate(atts):
    rearrangd_att = rearrange(
        att,
        "(b m) (H W) (n h w) -> b m H W n h w",
        n=6,
        m=4,
        H=H,
        W=W,
        h=hw[i][0],
        w=hw[i][1],
    )
    # b: batch n: num_camera, m: attention head number, H: downsampled bev height, W: downsampled bev width, h: downsampled camera height, w: downsampled camera weight
    # rearranged_atts.append(rearrangd_att)
    # b m H W n h w -> b m n h w H W
    cam_to_bev_att = rearrange(
        rearrangd_att,
        "b m H W n h w -> b m n h w H W",
    )
    cam_to_bev_atts.append(cam_to_bev_att)

cam_to_bev_att = cam_to_bev_atts[-1]
cam_to_bev_att = cam_to_bev_att[0]
mean_cam_to_bev_att = torch.mean(cam_to_bev_att, dim=0)
cam_idx = 1

y1, x1, y2, x2 = bboxes[-1]
bbox_att_map = mean_cam_to_bev_att[cam_idx, y1:y2, x1:x2].detach().cpu().numpy()
bbox_att_map = cv2.resize(
    bbox_att_map[0,0], dsize=(25, 25), interpolation=cv2.INTER_NEAREST
)
non_zero_min = np.min(bbox_att_map[bbox_att_map > 0])
_, thresh = cv2.threshold(bbox_att_map, non_zero_min, 255, cv2.THRESH_BINARY)
thresh = thresh.astype(np.uint8)
# thresh = np.expand_dims(thresh, axis=2)
# print(thresh.shape)
plt.imshow(thresh)
plt.show()
contours, _ = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
ex_cnt = contours[0]
bev_bbox = cv2.boundingRect(ex_cnt)
x, y, w, h = bev_bbox
bev_image = preds[0][0].detach().cpu().numpy()
bev_image = cv2.rectangle(bev_image, (x, y), (x + w, y + h), 1, 2)
plt.imshow(bev_image)
plt.imshow(bbox_att_map, alpha=bbox_att_map)
plt.show()
```

```python

def map (atts, cidx, x, y):
    atts = atts.cpu().numpy()

    max_x = 0
    max_y = 0
    max_att = 0

    for mx in range(25):
        for my in range(25):
            am = atts[mx][my][cidx][x][y]

            if am > max_att:
                max_att = am
                max_x = mx
                max_y = my

    return max_x, max_y, max_att
```
